```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
#df.drop('Car_Name',axis=1,inplace=True)
df['Car_Name']=df['Car_Name'].str.split().str.slice(0,1).str.join(' ')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True,keep="first")
```

```python
df=df.reset_index()
df=df.drop('index',axis=1)
```

```python
df['age']  = 2023 - df['Year'] 
df.drop(['Year'], axis = 1, inplace= True)
df.reset_index()
```

```python
df.head()
```

```python
df['Car_Name'].unique()
```

```python
temp=['sx4',  'wagon','s','ignis', '800','corolla', 'camry', 'land','UM', 'grand', 'eon', 'xcent','elantra','city', 'brio', 'jazz']
temp2=[]
for i in range(0,len(df)):
    if df.iloc[i]['Car_Name'] in temp:
        temp2.append(i)
# len(temp2)
df=df.drop(temp2)
df.reset_index(inplace=True)
df.head()
```

```python
df.drop(columns='index',axis=1,inplace=True)
```

```python
print(df["Fuel_Type"].unique())
print(df["Seller_Type"].unique())
print(df["Transmission"].unique())
print(df["Owner"].unique())
```

```python
X = df.drop("Selling_Price",axis=1)
y = df["Selling_Price"]
```

```python
X=pd.get_dummies(X,columns=["Car_Name","Fuel_Type","Seller_Type","Transmission","Owner"])
```

```python
pd.set_option("display.max_columns",None)

X.head()
```

```python
X.columns
```

```python
temp=['Car_Name_Activa',
       'Car_Name_Bajaj', 'Car_Name_Hero', 'Car_Name_Honda', 'Car_Name_Hyosung',
       'Car_Name_KTM', 'Car_Name_Mahindra', 'Car_Name_Royal',
       'Car_Name_Suzuki', 'Car_Name_TVS', 'Car_Name_Yamaha', 'Car_Name_alto',
       'Car_Name_amaze', 'Car_Name_baleno', 'Car_Name_ciaz', 'Car_Name_creta',
       'Car_Name_dzire', 'Car_Name_ertiga', 'Car_Name_etios',
       'Car_Name_fortuner', 'Car_Name_i10', 'Car_Name_i20', 'Car_Name_innova',
       'Car_Name_omni', 'Car_Name_ritz', 'Car_Name_swift', 'Car_Name_verna',
       'Car_Name_vitara', 'Fuel_Type_Diesel', 'Fuel_Type_Petrol',
       'Seller_Type_Dealer', 'Seller_Type_Individual',
       'Transmission_Automatic', 'Transmission_Manual', 'Owner_0', 'Owner_1']
for i in temp:
    X[i].replace({True:1,False:0},inplace=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
temp=X_train.columns

sc=StandardScaler()
temp2=X_train
sc.fit(temp2)
X_train_sc=sc.transform(temp2)
X_test_sc=sc.transform(X_test)
X_train_sc=pd.DataFrame(X_train_sc,columns=temp)
X_test_sc=pd.DataFrame(X_test_sc,columns=temp)

mms=MinMaxScaler()
temp2=X_train
mms.fit(temp2)
X_train_mms=mms.transform(temp2)
X_test_mms=mms.transform(X_test)
X_train_mms=pd.DataFrame(X_train_mms,columns=temp)
X_test_mms=pd.DataFrame(X_test_mms,columns=temp)
```

multiple linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

lr_sc = LinearRegression()
lr_sc.fit(X_train_sc, y_train)
y_pred_sc = lr_sc.predict(X_test_sc)

lr_mms = LinearRegression()
lr_mms.fit(X_train_mms, y_train)
y_pred_mms = lr_mms.predict(X_test_mms)

#Accuracy
from sklearn.metrics import r2_score
print("Basic => ",r2_score(y_test, y_pred))
print("StandardScalar => ",r2_score(y_test, y_pred_sc))
print("MinMaxScalar => ",r2_score(y_test, y_pred_mms))
```

ridge

```python
from sklearn.linear_model import Ridge
rd = Ridge()
rd.fit(X_train, y_train)
y_pred = rd.predict(X_test)

rd_sc = Ridge()
rd_sc.fit(X_train_sc, y_train)
y_pred_sc = rd_sc.predict(X_test_sc)

rd_mms = Ridge()
rd_mms.fit(X_train_mms, y_train)
y_pred_mms = rd_mms.predict(X_test_mms)

#Accuracy
from sklearn.metrics import r2_score
print("Basic => ",r2_score(y_test, y_pred))
print("StandardScalar => ",r2_score(y_test, y_pred_sc))
print("MinMaxScalar => ",r2_score(y_test, y_pred_mms))
```

lasso

```python
from sklearn.linear_model import Lasso
ls = Lasso()
ls.fit(X_train, y_train)
y_pred = ls.predict(X_test)

ls_sc = Lasso()
ls_sc.fit(X_train_sc, y_train)
y_pred_sc = ls_sc.predict(X_test_sc)

ls_mms = Lasso()
ls_mms.fit(X_train_mms, y_train)
y_pred_mms = ls_mms.predict(X_test_mms)

#Accuracy
from sklearn.metrics import r2_score
print("Basic => ",r2_score(y_test, y_pred))
print("StandardScalar => ",r2_score(y_test, y_pred_sc))
print("MinMaxScalar => ",r2_score(y_test, y_pred_mms))
```

Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
temp=['absolute_error', 'friedman_mse', 'poisson', 'squared_error']
for i in temp:
    dtr = DecisionTreeRegressor(criterion=i,random_state = 5)
    dtr.fit(X_train, y_train)

    y_pred_dtr = dtr.predict(X_test)

    print(i," -> ",r2_score(y_test, y_pred_dtr))

# so poisson is best, r2 score=0.8375043587814766
dtr = DecisionTreeRegressor(criterion='poisson',random_state = 5)
dtr.fit(X_train, y_train)

y_pred_dtr = dtr.predict(X_test)

print(r2_score(y_test, y_pred_dtr))
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
temp1=['absolute_error', 'friedman_mse', 'poisson', 'squared_error']
temp2=[100]
for i in range(200,601,200):
    temp2.append(i)

for i in temp1:
    for j in temp2:
        rf = RandomForestRegressor(n_estimators = j,criterion=i,random_state = 5)
        rf.fit(X_train, y_train)

        y_pred_rf = rf.predict(X_test)

        print(i," and ",j," -> ",r2_score(y_test, y_pred_rf))

# so we got great r2 score with friedman_mse  and  100  ->  0.8807247931506951
rf = RandomForestRegressor(n_estimators = j,criterion=i,random_state = 5)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

r2_score(y_test, y_pred_rf)
```

K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsRegressor

for i in range(3,52,2):
    knn=KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train, y_train)

    y_pred_knn = knn.predict(X_test)

    print(i," -> ",r2_score(y_test, y_pred_knn))

# so knn works pretty bad here
```
